```python
import requests
import json
link = "https://raw.githubusercontent.com/MarvinJacobPMumar/kamia-mumar-grade8-library-Repo/refs/heads/main/books.json"
response = requests.get(link)
books = response.json()
print(books)
```

```python
import firebase_admin
from firebase_admin import credentials, db

credential = credentials.Certificate("/content/firebase_key.json")
firebase_admin.initialize_app(credential, { "databaseURL" : "https://kamia-mumar-grade8-library-db-default-rtdb.asia-southeast1.firebasedatabase.app/"})
print("Succesfully connected to Firebase.")

```

```python
import json

data = books
print("Loaded JSON file")
reference = db.reference("books")
for book in data["books"]:
  reference.child(str(book["id"])).set(book)
print("Data succesfully uploaded.")
```

```python
books_data = reference.get()
print(f"Data: {books_data} \n")

print("Books in library:", end = " ")
for i, book in enumerate(books_data.values()):
  if i + 1 == len(books_data.values()):
    print(f"and {book["title"]} \n")
  else:
    print(book["title"], end = ", ")


booksinlibrary = 0
print("Availability for each book:")
for book in books_data.values():
  details = book["details"]
  print(f"  {book["title"]}:")
  print(f"    {details["available"]}/{details["copies"]} copies")
  booksinlibrary += details["copies"]

print(f"\nTotal books in the library: {booksinlibrary}")



```
